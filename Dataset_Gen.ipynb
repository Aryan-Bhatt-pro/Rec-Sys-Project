```python
import random
import csv

# Generating synthetic historical prices gradually increasing from 150 to 180 with occasional falls
historical_prices = []
current_price = 57.4
for _ in range(80):
    
    current_price += random.uniform(-0.6, 1.2)
    current_price = max(min(current_price, 70.4), 57)   # Ensure prices stay within 48 to 52 range
    historical_prices.append(current_price)


# Synthetic dataset generation
dataset = []

c = 0
for price in historical_prices[:-1]:
    
    error = random.uniform(0, 0.04)
    expected_value = historical_prices[c + 1] + ((historical_prices[c + 1] - price) * error * random.choice([-1,1]))
    error = error*price
    percentage_portfolio = random.uniform(0.5, 5.0)
    percentage_sector = random.uniform(5.0, 20.0)
    sector = 'IT'
    action = ""
    if random.random() > 0.8:
        action = random.choice(['Buy', 'Sell', 'Keep'])
    else:
        if ((expected_value - price) > error):
            action = 'Buy'
        elif ((price - expected_value) > error):
            action = 'Sell'
        else:
            action = 'Keep'

    row = [price, expected_value, error, percentage_portfolio, percentage_sector, sector, action]
    dataset.append(row)
    c += 1
print(dataset)
# Writing the dataset to a CSV file
with open('tgt_stock_trades.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Current Value of Stock', 'Expected Value of Stock', 'Error in expected value of stock',
                     'Percentage of portfolio', 'Percentage of sector', 'Sector of stock', 'Buy/Sell/Keep'])
    writer.writerows(dataset)

print("Dataset generated and saved to apple_stock_trades.csv")
```
